```python
ECN_EPI1 = w_EPI1.T[6:13, 6:13]
ECN_lower_EPI1 = np.tril(ECN_EPI1,-1)
ECN_flat_EPI1 = squareform(ECN_EPI1, 'tovector', checks=False)
ECN_mean_EPI1 = np.mean(ECN_flat_EPI1)
ECN_EPI1_SD = np.std(ECN_flat_EPI1)

ECN_EPI2 = w_EPI2.T[6:13, 6:13]
ECN_lower_EPI2 = np.tril(ECN_EPI2,-1)
ECN_flat_EPI2 = squareform(ECN_EPI2, 'tovector', checks=False)
ECN_mean_EPI2 = np.mean(ECN_flat_EPI2)
ECN_EPI2_SD = np.std(ECN_flat_EPI2)

print(ECN_mean_EPI1)
print(ECN_mean_EPI2)
```

```python
SN_EPI1 = w_EPI1.T[13:, 13:]
SN_lower_EPI1 = np.tril(SN_EPI1,-1)
SN_flat_EPI1 = squareform(SN_EPI1, 'tovector', checks=False)
SN_mean_EPI1 = np.mean(SN_flat_EPI1)
SN_EPI1_SD = np.std(SN_flat_EPI1)

SN_EPI2 = w_EPI2.T[13:, 13:]
SN_lower_EPI2 = np.tril(SN_EPI2,-1)
SN_flat_EPI2 = squareform(SN_EPI2, 'tovector', checks=False)
SN_mean_EPI2 = np.mean(SN_flat_EPI2)
SN_EPI2_SD = np.std(SN_flat_EPI2)

print(SN_mean_EPI1)
print(SN_mean_EPI2)
```
