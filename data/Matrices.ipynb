```python
DMN_EPI1 = w_EPI1.T[0:6, 0:6]
DMN_lower_EPI1 = np.tril(DMN_EPI1,-1)
total = 0
for i in DMN_lower_EPI1:
    total += i
DMN_mean_EPI1 = np.sum(total)/15

DMN_EPI2 = w_EPI2.T[0:6, 0:6]
DMN_lower_EPI2 = np.tril(DMN_EPI2,-1)
total = 0
for i in DMN_lower_EPI2:
    total += i
DMN_mean_EPI2 = np.sum(total)/15

print(DMN_mean_EPI1)
print(DMN_mean_EPI2)
```

```python
ECN_EPI1 = w_EPI1.T[6:13, 6:13]
ECN_lower_EPI1 = np.tril(ECN_EPI1,-1)
total = 0
for i in ECN_lower_EPI1:
    total += i
ECN_mean_EPI1 = np.sum(total)/21


ECN_EPI2 = w_EPI2.T[6:13, 6:13]
ECN_lower_EPI2 = np.tril(ECN_EPI2,-1)
total = 0
for i in ECN_lower_EPI2:
    total += i
ECN_mean_EPI2 = np.sum(total)/21

print(ECN_mean_EPI1)
print(ECN_mean_EPI2)
```

```python
SN_EPI1 = w_EPI1.T[13:, 13:]
SN_lower_EPI1 = np.tril(SN_EPI1,-1)
total = 0
for i in SN_lower_EPI1:
    total += i
SN_mean_EPI1 = np.sum(total)/15

SN_EPI2 = w_EPI2.T[13:, 13:]
SN_lower_EPI2 = np.tril(SN_EPI2,-1)
total = 0
for i in SN_lower_EPI2:
    total += i
SN_mean_EPI2 = np.sum(total)/15

print(SN_mean_EPI1)
print(SN_mean_EPI2)
```

```python
EPI1_NWmeans={"DMN": DMN_mean_EPI1, "ECN": ECN_mean_EPI1 , "SN": SN_mean_EPI1}
EPI2_NWmeans ={"DMN": DMN_mean_EPI2, "ECN": ECN_mean_EPI2 , "SN": SN_mean_EPI2}
print(EPI1_NWmeans)
print(EPI2_NWmeans)
```

```python
DMN_flat_EPI1 = squareform(DMN_EPI1, 'tovector', checks=False)
DMN_flat_EPI2 = squareform(DMN_EPI2, 'tovector', checks=False)
ECN_flat_EPI1 = squareform(ECN_EPI1, 'tovector', checks=False)
ECN_flat_EPI2 = squareform(ECN_EPI2, 'tovector', checks=False)
SN_flat_EPI1 = squareform(SN_EPI1, 'tovector', checks=False)
SN_flat_EPI2 = squareform(SN_EPI2, 'tovector', checks=False)
```

```python
print(SN_flat_EPI2)
```

```python
DMN_EPI1_SD = np.std(DMN_flat_EPI1)
DMN_EPI2_SD = np.std(DMN_flat_EPI2)
ECN_EPI1_SD = np.std(ECN_flat_EPI1)
ECN_EPI2_SD = np.std(ECN_flat_EPI2)
SN_EPI1_SD = np.std(SN_flat_EPI1)
SN_EPI2_SD = np.std(SN_flat_EPI2)
```

```python
EPI1_NWSD={"DMN": DMN_EPI1_SD, "ECN": ECN_EPI1_SD , "SN": SN_EPI1_SD}
EPI2_NWSD={"DMN": DMN_EPI2_SD, "ECN": ECN_EPI2_SD , "SN": SN_EPI2_SD}
print(EPI1_NWSD)
print(EPI2_NWSD)
```
